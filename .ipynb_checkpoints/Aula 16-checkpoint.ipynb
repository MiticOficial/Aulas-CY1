```python
#HTML

'''
As páginas na internet são estruturadas através de HTML, 
uma linguagem de marcação que especifica como as informações de sites
devem se comportar nos navegadores web.

A linguagem HTML é baseada em uma sintaxe de tags. 
As tags definem comandos que estruturam informações em sites, 
e são delimitadas pelo sinal de menor que “<” e maior que “>”.

maioria das tags são apresentadas em pares (tag de abertura e tag de fechamento).

As tags que não possuem uma tag de abertura, são chamadas órfãs. 
Todo conteúdo entre as tags de abertura e fechamento de um comando, 
pertencem a ele, inclusive outras tags.

Ex de um sintaxe basica em html:

<html>
    <head>
        <title>Ctrl+Play</title>
    </head>
    <body>
        Escola de programação e robotica!
    </body>
</html>

A maior parte dessas tags são geradas automaticamente com poucos comandos, como o <!DOCTYPE html> 
'''

#TAGS

'''
As tags <head> informam dados importantes para os desenvolvedores, 
como a definição de codificação de caracteres e importação de arquivos externos. 
Enquanto as tags title e body definem, respectivamente, o título e o conteúdo principal do site. 
Assim, tudo o que é definido entre as tags head não estará visível na página e entre as tags body o que é visível

As tags podem possuir atributos, os mais importantes para o processo de coleta de dados web
(também conhecido como Web Scraping) são os chamados seletores. 
Existem dois seletores para identificação de tags HTML, o id e o class. 
O id é um identificador único de elementos de uma página web,
já o seletor class é utilizado para definir um padrão de elementos,
agrupando vários elementos com um determinado atributo.
'''
```

```python
#Trabalhando com o beautiful Soup

'''
Para operar sobre o conteúdo HTML de uma página web usando Beautiful Soup,
você primeiramente deve instanciar um objeto Request

no caso do exemplo abaixo seria o:  import requests as r

em seguida instanciá-la, passando o conteúdo em texto para seu construtor(.get())
e armazenando o retorno numa variável qualquer, como no exemplo da imagem a seguir.

Com isso feito, podemos usar varios tipos de metodos para recuperar dados de uma pagina em HTML,
assim como fazemos no exemplo abaixo.

Importante: Vale lembrar que sempre que você tentar acessar uma tag diretamente pelo seu nome, 
será retornada a primeira tag encontrada com este nome.
'''

import requests as r # Importando a biblioteca requests, referenciada como a letra r
from bs4 import BeautifulSoup # da biblioteca beautifull soup 4 eu vou importar a classe BeautifulSoup

#Tenta acessar o site do UOL
try:
    resultado = r.get("https://www.uol.com.br/") #Instanciando o objeto request (r) na variavel resultado

#Se existir algum erro ele deve mostrar "Erro: nome do erro"
except Exception as erro:
    print("erro: ", erro)

#Se naõ existir nenhum erro ele vai extrair o conteudo da pagina como texto.
else:
    resposta = resultado.text
    soup = BeautifulSoup(resposta, "html.parser") # Essa linha converte o HTML em uma estrutura navegável, permitindo extrair e manipular informações facilmente.

    #Acessa a tag <title>(titulo) do html e exibe o texto que está dentro dela
    print(soup.title.string)

print(resposta)#Mostrando o resultado caso puxemos só o resposta sem prettify
```

```python
'''
Outra maneira de recuperar diretamente um elemento pelo seu nome é usando o método find(), 
que aceita o nome do atributo ou um atributo com determinado valor e uma class, 
que é um subtipo de atributo.

No exemplo a seguir, é recuperado os dados de um elemento h3 de um subtipo do site do uol.
'''


#Importando as bibliotecas
import requests as r
from bs4 import BeautifulSoup

#Tenta acessar o site do UOL
try:
    resultado = r.get("https://www.uol.com.br/")

#Se existir algum erro ele deve mostrar "Erro: nome do erro"
except Exception as erro:
    print("erro: ", erro)


else:
    #O HTML da página é transformado em um objeto soup para facilitar a navegação.
    resposta = resultado.text
    soup = BeautifulSoup(resposta, "html.parser")
    
    #Busca a primeira tag <h3> que tenha essa classe CSS.
    #o prettify() formata o html de forma legível antes de exibir
    print(soup.find("h3", class_="title__element headlineMain__title").prettify())


'''
O metodo .prettify() serve para organizar a estrutura html do jeito mais tradicional em que ela é escrita
facilitando a leitura.
'''
    
```

```python
'''
Depois de analisar a estrutura do site, vamos criar um programa para verificar a noticia de alguns jogos!
Para isso, é necessario uma lista com os nomes do jogos que iremos buscar noticias no site, no caso, eu usei
LOL, Valorant, FIFA e Fortnite, mas você pode alterar ou adicionar itens nessa lista!
'''

import requests #Importei a biblioteca requests
from bs4 import BeautifulSoup #importei da biblioteca beautiful soup 4 a classe BeautifulSoup

link = "https://www.uol.com.br/start/esport/" #O link da pagina que eu quero acessar 
pagina = requests.get(link) #Colocando as informações da pagina na minha variavel pagina

soup = BeautifulSoup(pagina.content, "html.parser") #pegando o conteudo da pagina com o beautifulsoup

lista = ["lol","Valorant","fifa","fortnite"] #Lista de jogos que iremos procurar noticias

for paragrafo in soup.find_all("body"):
    for palavra in lista:
        for paragrafo_str in paragrafo.stripped_strings:
            if palavra.upper() in str(paragrafo_str).upper():
                print("Noticia sobre: ", palavra.upper(), '\n', paragrafo_str, '\n')
                break

'''
E nessa estrutura de repetição for eu fui buscando entre os paragrafos encontrados na tag <body> por palavras
que estavam na lista (lol, valorant, fifa, fortnite). Em 'for paragrafo_str in paragrafo.stripped_strings'
ele retorna os textos em html removendo os espaços extras. E no 'if palavra.upper() in str(paragrafo_str).upper():'
se existirem alguma noticia sobre alguma das palavras chave, ele irá mostrar "Noticia sobre: " a palavra chave da lista(lol, valorant, fifa e fortnite)
e a manchete da noticia!
'''
```
