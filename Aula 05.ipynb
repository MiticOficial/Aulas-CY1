```python
#Estruturas de repetição;

'''
Estruturas de repetição no python servem para alterar o fluxo de execução do
código Python, pois tornam possivel a repetição de trechos especificos até que
uma condição seja atingida.

existem dois tipos de estruturas de Repetição no Python, o For e o While.
'''

itens_de_mercado = ["Arroz","Feijão","Macarrão","Vinagre"]

for item in itens_de_mercado:
    print("Os itens são: ", item)

#Estrutura de repetição "For";

'''
A estrutura for atua como um iterador ou seja um objeto que percorre itens que
estão em uma sequência e retorna os dados desses itens de maneira sucessiva,
um elemento de cada vez. É possivel iterar principalmente sobre strings e listas.

formula para usar a estrutura de repetição FOR:

for i in objeto:
    codigo que vai repetir.

*esse i significa item, mas para abreviar usamos i.67

Então, em resumo, o for é um jeito de “passear” por uma lista de itens ou um conjunto de números,
fazendo algo para cada um deles.
'''

caes = ['Cachorro1','Cachorro2','Cachorro3','Cachorro4']
print(caes)

for cao in caes:
    print(cao)
    
print("-----------------------------")

numeros = list(range(1,11))
for num in numeros:
    print(num)

#Para gerar uma lista de numeros mais rapidamente podemos utilizar o range junto com o for
print("-----------------------------")
for num in range(1,11):
    print(num)

print("-----------------------------")

'''
O nome do item pode ser referenciado dentro do seu loop para, por exemplo, 
utilizar-se do if para verificar se há números pares em uma lista.
'''

for num in range(1,11):
    if(num %2 ==0):
        print(num)
print("-----------------------------")
'''
Podemos usar também com Strings, porem como strings são sequencias de caracteres
você terá acesso a essa sequencia!
'''

#Iterando sobre uma String
for letra in "Ctrl+Play":
    print(letra)
```

```python
#Estrutura de repetição While

'''
A estrutura while é uma das formas mais gerais de executar iterações.
Basicamente ela executa um trecho de código enquanto uma condição for
verdadeira!

a formula para ela é:

while condição:
    codigo que vai repetir
'''

#Exemplo


soma = 0 
numeros = range(1,1000)
i = 0

while soma < 100:
    soma += numeros[i]
    i+=1
print(soma)

#Podemos adicionar comandos para caso a condição se tornar falso com o else:

x = 0
while x < 5:
    print('valor de x: ', x)
    print("x ainda é menor que 5, vamos somar 1")
    x =+ 1
else:
    print("Fim")
```

```python

#DESAFIOS:


def fibonacci_sequence(n):
    sequence = [0, 1]
    for i in range(2, n):
        next_value = sequence[i - 1] + sequence[i - 2]
        sequence.append(next_value)
    return sequence[:n]

# Leitura de N com condição de que N < 46
N = int(input("Digite um valor para N (N < 46): "))
if 0 < N < 46:
    resultado = fibonacci_sequence(N)
    print(" ".join(map(str, resultado)))
else:
    print("Por favor, insira um valor de N entre 1 e 45.")
```
