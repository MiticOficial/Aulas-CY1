```python
# Conhecendo o Git e o Github

#O que é o controle de versão de código?

'''
Você já deve ter criado pastas e pastas com os mesmos arquivos, 
só tendo uma pequena modificação, para caso você precise voltar a uma versão anterior, 
fique “fácil” de executar essa ação. O problema é que com projetos grandes, 
isso acaba ocupando muito espaço no seu HD 
e a tarefa de achar um arquivo correto se torna algo difícil
- qual a diferença entre os arquivos “Trabalho1”, “TrabalhoFim”, “TrabalhoAgoraVai”? 

Para resolver esses problemas, surgiu o Controle de Versão! 
Um sistema para gerenciamento de diferentes versões de documentos, 
que consegue “enxergar” todas as modificações que você fez em um projeto e as armazena em forma de versões. 
Com isso você consegue navegar e escolher em qual delas deseja trabalhar sem ter tanta dificuldade.

Além disso, ter um maior controle desses arquivos permite de forma muito mais facil, 
trabalhar em conjunto com outros DEVs. 

Existem diversos sistemas de versionamento de código, mas o mais utilizado e o que veremos hoje é o GIT.
'''

#Primeiros passos com o Git

'''
O git é o principal sistema de versionamento de código utilizado no meio da programação,
muito por ter seu código base aberto e gratuito. Com o git, podemos desenvolver projetos colaborativos(ou não),
de maneira organizada e rastreável. 

'''

#OLHAR SITUAÇÕES E SOLUÇÕES DA PAGINA GIT DA APOSTILA!

#Conhecendo alguns comandos importantes do CMD (Ou prompt de comando)

'''
#CD - O comando CD é usado para navegar entre os diretorios do computador, então por exemplo,
se eu quero acessar uma pasta que está dentro da area de trabalho, esse seria o comando:

cd C:\Users\CtrlPlay\Desktop\Nomedapasta

#Mkdir - O comando Mkdir (que eu leio como make diretory) faz exatamente o que o nome sugere, ele cria uma pasta
(ou diretorio) no lugar aonde você está no sistema:

mkdir c:\Users\CtrlPlay\Desktop\Programa\NovaPasta

#dir - O comando dir lista todos os arquivos que existem dentro da pasta em que você está no momento
'''

#Como funciona o Git?

'''
Neste momento, você já tem tudo pronto para controlar a versão dos seus códigos, 
porém precisa conhecer dois conceitos muito importantes para quem utiliza Git. 
Todo sistema de controle de versão é composto por duas partes: 
o “repositorio remoto” e o “repositorio local”.

O repositorio remoto armazena todo o histórico e evolução do projeto em um outro local (servidor), 
enquanto que o repositorio local é uma cópia individual e isolada dos arquivos do projeto presente na sua maquina, e é nele que você trabalha

'''

#Sincronização

'''
Para sincronizar os repositorios locais e remotos, devemos utilizar os comandos Add, commit e update

- Comando Add: Adiciona todas as mudanças feitas do repositorio local (do arquivo que você está editando),
na stanging area, uma região semelhante a um embarque nos aeroportos, onde tudo é checado e pode ser modificado
antes de ser enviado!

- Comando commit: é utilizado quando é preciso enviar mudanças feitas no repositorio local
para o repositório remoto

- Comando update: Já o update é utilizado para retornar um arquivo do repositorio remoto para o repositorio local
'''


#Então a ordem para a utilização dos comandos é a seguinte:

'''

C:\Users\CtrlPlay\Desktop\Nomedapasta git add . (O ponto é necessario!)

Após esse comando os seus arquivos estão na stanging area (aquela area parecida com um embarque de um aeroporto)

git commit -m “Meu primeiro commit”

Após esse comando o arquivo está no seu repositorio local
'''
```
