```python
#Estruturas condicionais:

#Operadores de comparação:

'''
Antes de falar sobre Estruturas condicionais, precisamos falar um pouco sobre operadores
de comparação, uma parte crucial para essas estruturas!

Operadores de comparação são utilizados para verificar variaveis e expressões e então produzir um valor
BOOLEANO(True ou False)
'''

#Esses são alguns dos operadores:

#Igual: verifica se os dois operandos são iguais, se sim retorna True, se não retorna false!


10 == 10 #True
#Não se esqueça que podemos também comparar variaveis!

numero1 = 10
numero2 = 10

numero1 == numero2 
#Resultado: True

#Diferente: verifica se um dos operandos é diferente do outro, se sim retorna True, se não retorna false!
10 != 25 #True

#Maior: verifica se o operando esquerdo é maior que o operando direito, se sim retorna True, se não retorna false!
10 > 9 #True

#Menor: verifica se o operando esquerdo é menor que o operando direito, se sim retorna True, se não retorna false!
10 < 11 #True

#Maior igual: verifica se o operando esquerdo é maior OU igual que o operando direito, se sim retorna True, se não retorna false!
10 >= 10 #True
#Note que 10 não é maior que 10, mas 10 é igual a dez, então uma das condições se cumpre, logo o resultado é verdadeiro

#Menor igual: verifica se o operando esquerdo é Menor OU igual que o operando direito, se sim retorna True, se não retorna false!
10 <= 55 #True

#Operadores de comparação em cadeia:

'''
O Python tem uma caracteristica muito interessante, ela permite encadear comparações multiplas para realiza
varios tipos de testes.
'''

#Exemplo: 

1 < 2 < 3 #Todas as condições são verdadeiras, logo retorna True

#Comandos and e or:

#Comando and: 
'''
O comando AND é usado para garantir que as duas verificações tenham que ser verdadeiras
para que a verificação total seja verdadeira!

ou seja (verificação verdadeira) and (verificação verdadeira) = True
'''
#Exemplo and:
1 > 0 and 120 < 100 #False

#Comando or: 
'''
O comando OR é usado para verificar se UMA das verificações for verdade, logo a verificação total é verdadeira!

ou seja (verificação verdadeira) or (verificação falsa) = True
'''

#Exemplo or:
1=523 or 10<20 #True
```

```python
#Estrutura condicional:
'''
Estruturas condicionais são estruturas simples ou compostas que executam um grupo de ações 
quando determinadas condições (definidas por você) são atendidas ou não.
'''

#Estrutura condicional simples:

'''
Uma estrutura condicional simples faz executar um ou vários comandos SE a condição for verdadeira. 
Se a condição for falsa, a estrutura é finalizada sem executar os comandos.
'''

#Exemplo:

nota = 7
if nota >= 6:
    print("Aprovado")

#Regras aplicadas no exemplo acima:
'''
Para criar uma estrutura condicional simples, precisamos seguir algumas regras.
    1- No fim das verificações da condição, para indicar que elas já terminaram, 
    deve-se usar dois pontos (:);
    2- Os códigos que serão executados caso as condições sejam verdadeiras devem ser indentados (com um recuo maior)
'''
```

```python
#Estrutura condicional composta:
'''
A estrutura condicional composta segue a mesma lógica da estrutura condicional simples: 
executar comandos a partir da verificação de uma condição.
A diferença que existe na estrutura composta é que ela permite uma nova bifurcação, 
uma nova alternativa dentro do seu código, que é executar comandos se a condição for verificada como falsa
'''

#Exemplo(Com if e else apenas):

nota = 7
if nota >= 5 :
    print("Passou de ano")
else: 
    print("Não passou")

#Exemplo(Com if, elif e else):

if nota >= 6:
    print("Aprovado!")
elif 4 > nota < 6:
    print("recuperação")
else:
    print("Reprovado!")

#Use input(), variaveis e estruturas condicionais


```
