```python
#HTML

'''
As páginas na internet são estruturadas através de HTML, 
uma linguagem de marcação que especifica como as informações de sites
devem se comportar nos navegadores web.

A linguagem HTML é baseada em uma sintaxe de tags. 
As tags definem comandos que estruturam informações em sites, 
e são delimitadas pelo sinal de menor que “<” e maior que “>”.

maioria das tags são apresentadas em pares (tag de abertura e tag de fechamento).

As tags que não possuem uma tag de abertura, são chamadas órfãs. 
Todo conteúdo entre as tags de abertura e fechamento de um comando, 
pertencem a ele, inclusive outras tags.

Ex de um sintaxe basica em html:

<html>
    <head>
        <title>Ctrl+Play</title>
    </head>
    <body>
        Escola de programação e robotica!
    </body>
</html>

A maior parte dessas tags são geradas automaticamente com poucos comandos, como o <!DOCTYPE html> 
'''

#TAGS

'''
As tags <head> informam dados importantes para os desenvolvedores, 
como a definição de codificação de caracteres e importação de arquivos externos. 
Enquanto as tags title e body definem, respectivamente, o título e o conteúdo principal do site. 
Assim, tudo o que é definido entre as tags head não estará visível na página e entre as tags body o que é visível

As tags podem possuir atributos, os mais importantes para o processo de coleta de dados web
(também conhecido como Web Scraping) são os chamados seletores. 
Existem dois seletores para identificação de tags HTML, o id e o class. 
O id é um identificador único de elementos de uma página web,
já o seletor class é utilizado para definir um padrão de elementos,
agrupando vários elementos com um determinado atributo.
'''
```

```python
#Trabalhando com o beautiful Soup

'''



'''


import requests as r
from bs4 import BeautifulSoup

try:
    resultado = r.get("https://www.uol.com.br/")
except Exception as erro:
    print("erro: ", erro)

else:
    resposta = resultado.text
    soup = BeautifulSoup(resposta, "html.parser")
    print(soup.title.string)
```

```python
import requests as r
from bs4 import BeautifulSoup

try:
    resultado = r.get("https://www.uol.com.br/")
except Exception as erro:
    print("erro: ", erro)

else:
    resposta = resultado.text
    soup = BeautifulSoup(resposta, "html.parser")
    print(soup.find("h3", class_="title__element headlineMain__title").prettify())
```

```python
'''



'''

import requests
from bs4 import BeautifulSoup

url = "https://www.uol.com.br/esporte/"
page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")

lista = ["lol","Valorant","fifa","fortnite"]

for paragrafo in soup.find_all("body"):
    for palavra in lista:
        for paragrafo_str in paragrafo.stripped_strings:
            if palavra.upper() in str(paragrafo_str).upper():
                print("Noticia sobre: ", palavra.upper(), '\n', paragrafo_str, '\n')
                break
                
```
